```python
import matplotlib.pyplot as plt
import numpy as np

from interpSph import *
filename = "./DataForInterp2.csv"
```

```python
data, theta, phi = loadFile(filename)
theta_r = np.linspace(theta.min(), theta.max(), 180)
phi_r = np.linspace(phi.min(), phi.max(), 360)
PHI, THETA = np.meshgrid(phi, theta)
PHI_R, THETA_R = np.meshgrid(phi_r, theta_r)
plt.figure('original')
plt.pcolormesh(PHI * 180 / np.pi, THETA * 180 / np.pi, data, cmap=mpl.cm.jet, alpha=0.7)
plt.gca().invert_yaxis()
```

```python
bool = ~np.isnan(data)
interp = LinearNDInterpolator(list(zip(THETA[bool].ravel(),PHI[bool].ravel())), data[bool].ravel())
data = interp(THETA_R, PHI_R)
plt.figure('LinearNDInterpolator')
plt.pcolormesh(PHI_R * 180 / np.pi, THETA_R * 180 / np.pi, data, cmap=mpl.cm.jet, alpha=0.7)
plt.gca().invert_yaxis()
```

```python
data, theta, phi = loadFile(filename)
theta_r = np.linspace(theta.min(), theta.max(), 180)
phi_r = np.linspace(phi.min(), phi.max(), 360)
PHI, THETA = np.meshgrid(phi, theta)
PHI_R, THETA_R = np.meshgrid(phi_r, theta_r)

data, theta, phi = fillNaN(data, theta, phi)
plt.figure('fill NaN')
plt.pcolormesh(PHI * 180 / np.pi, THETA * 180 / np.pi, data, cmap=mpl.cm.jet, alpha=0.7)
plt.gca().invert_yaxis()
```

```python
interp = RegularGridInterpolator((theta, phi), data)

data = interp((THETA_R, PHI_R))
plt.figure('RegularGridInterpolator')
plt.pcolormesh(PHI_R * 180 / np.pi, THETA_R * 180 / np.pi, data, cmap=mpl.cm.jet, alpha=0.7)
plt.gca().invert_yaxis()
```
